```python
# Printing the structure of thee board
def print_board(board):
    print('   |   |')
    print(' ' + board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('   |   |')
```

```python
# Main Body
def is_winner(board, player):
    return (
        (board[7] == player and board[8] == player and board[9] == player) or
        (board[4] == player and board[5] == player and board[6] == player) or
        (board[1] == player and board[2] == player and board[3] == player) or
        (board[7] == player and board[4] == player and board[1] == player) or
        (board[8] == player and board[5] == player and board[2] == player) or
        (board[9] == player and board[6] == player and board[3] == player) or
        (board[7] == player and board[5] == player and board[3] == player) or
        (board[9] == player and board[5] == player and board[1] == player)
    )
```

```python
# Checking the positions available
def is_board_full(board):
    for i in range(1, 10):
        if board[i] == ' ':
            return False
    return True
```

```python
# Working
def play_game():
    board = [' '] * 10
    player1 = 'X'
    player2 = 'O'
    current_player = player1

    print('Welcome to Tic Tac Toe!')
    print_board(board)

    while True:
        position = int(input('\n Player ' + current_player + ', choose your position (1-9): '))
        if board[position] == ' ':
            board[position] = current_player
            print_board(board)

            if is_winner(board, current_player):
                print('Congratulations, Player ' + current_player + ' has won!')
                break

            if is_board_full(board):
                print('The game is a tie!')
                break

            current_player = player2 if current_player == player1 else player1
        else:
            print('That position is already taken, please choose another one.')
```

```python
play_game()
```
